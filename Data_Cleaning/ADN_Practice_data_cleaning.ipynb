```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Data Analyst/Assignment/dataset.txt"
                       ,header = None,delimiter = ' ')

data.columns = ['Province_State', 'Country_Region', 'Lat','Long','Date','Confirmed','Deaths']
  
data.to_csv('/content/drive/MyDrive/Data Analyst/Assignment/dataset.csv', 
                index = None)
```

```python
data
```

```python
data = data.drop(index=0)
data
```

```python
for col in data.columns:
    print(col)
```

```python
province_state_0 = data[data['Province_State'].str.contains('0')]
province_state_0
```

```python
rows,col = province_state_0.shape
print("Number of dropped rows: ",rows)
```

```python
data_without_province_state_0 = data.drop(data[data['Province_State'].str.contains('0')].index)
data_without_province_state_0
```

```python
data_without_province_state_0.info()
```

```python

data.isnull().sum()
```

```python
total_Null = 1+4+5
total_Null
```

```python
final_data_after_cleaning = data_without_province_state_0.fillna(0)
final_data_after_cleaning.isnull().sum()
```

```python
final_data_after_cleaning['Confirmed'] = final_data_after_cleaning['Confirmed'].astype(float)
final_data_after_cleaning['Deaths'] = final_data_after_cleaning['Deaths'].astype(float)
final_data_after_cleaning['Lat'] = final_data_after_cleaning['Lat'].astype(float)
final_data_after_cleaning['Long'] = final_data_after_cleaning['Long'].astype(float)
final_data_after_cleaning.info()
```

```python
final_data_after_cleaning
```

```python
import matplotlib.pyplot as plt

x = np.array(final_data_after_cleaning["Deaths"])
y = np.array(final_data_after_cleaning["Confirmed"])

plt.scatter(x, y)
plt.show()
```

```python
plt.plot(y, linestyle = 'dotted')
plt.show()
```

```python
plt.hist([x, y],  color=['blue', 'red'], label = ['Deaths', 'Confirmed'], density='True')
```

```python
final_data_after_cleaning = final_data_after_cleaning.rename(columns={'Lat':'LAT','Long':'LON'})
final_data_after_cleaning.head(5)
```
